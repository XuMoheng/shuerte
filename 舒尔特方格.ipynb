```python
import pyautogui
from PIL import Image
from matplotlib import pyplot as plt
import time
from threading import Thread
```

```python
target=[]
target.append(0)
for i in range(1,17):
    img=Image.open('./number/'+str(i)+'.png')
    target.append(img)
go_img=Image.open('./number/go.png')
plt.imshow(go_img)
# plt.imshow(target[2])
```

```python
center={}
def func(num:int):
    try:
        left, top, width, height = pyautogui.locateOnScreen(image=target[num],confidence=0.9,grayscale=True)  # 寻找刚才保存点赞手势图片
        tmp_center = pyautogui.center((left, top, width, height))  # 寻找图片的中心
        center[num]=tmp_center
    except TypeError:
        print('没有找到', i)
```

```python
# for i in range(3,0,-1):
#     print('倒计时: ',i)
#     time.sleep(1)
try:
    left, top, width, height = pyautogui.locateOnScreen(image=go_img,confidence=0.9,grayscale=True)  # 寻找刚才保存点赞手势图片
    tmp_center = pyautogui.center((left, top, width, height))  # 寻找图片的中心
    pyautogui.click(tmp_center)
    print('开始')
except TypeError:
    print('开始失败')

threads=[]
center={}
for i in range(1,17):
    t=Thread(target=func,args=(i,))
    t.start()
    threads.append(t)

for i in range(1,17):
    pyautogui.click(center[i])
    # print('成功点击 ', i)

for t in threads:
    t.join()
```
